# Matplotlib Practice

```python
import matplotlib.pyplot as plt # Importing the matplotlib library
```

```python
# Matplotlib plot BASICS:
#------------------------
# line customization. markerfacecolor also called mfc, markeredgecolor also called mec.
line_style = dict(marker="*", markersize=10, markerfacecolor="blue", markeredgecolor="blue",
               linestyle='dashed', linewidth=2, color="red") 

x = [2023, 2024, 2025, 2026]
y = [15, 25, 30, 20]
plt.plot(x, y)

# Numpy arrays are faster than python lists.
import numpy as np
x = np.array([2023, 2024, 2025, 2026]) 
y1 = np.array([25, 30, 40, 30])
y2 = np.array([17, 23, 38, 5])
plt.plot(x,y1, **line_style) # ** unpacks the dictionary

# Adding a title to the plot
plt.title("Class Size", fontsize=20,
                        family="Arial",
                        fontweight="bold",
                        color="blue") 

# Adding a label in the x axis to the plot
plt.xlabel("Year", fontsize=15,
                    family="Arial",
                    fontweight="bold",
                    color="blue")

# Adding a label in the y axis to the plot
plt.ylabel("Students", fontsize=15,
                    family="Arial",
                    fontweight="bold",
                    color="blue")

# All the values in the x and y axis are now blue.
plt.tick_params(axis="both",
                colors="blue") 

# Sets the x axis ticks in a plot, now it will only show what the x array contains.
plt.xticks(x) 

# Grids.
plt.grid(axis="both", 
         linewidth="2",
         color='lightgray',
         linestyle='dashed') 

# shows the plot
plt.show() 
```

```python
# Bar Charts: 
# A bar chart is a graphical display that uses rectangular bars to compare the frequency or value of discrete categories of data. 
# The length or height of each bar is proportional to the value it represents, making it easy to visually compare different groups
#------------

categories = np.array(["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, values, color="skyblue") # call barh for a horizontal barchart
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
# Pie Charts:
# A pie chart is a circular statistical graph divided into slices, where each slice represents a numerical proportion of the whole. 
# The entire circle represents 100% of the data, and the size of each slice is proportional the quantity it represents. 
# Pie charts are useful for showing the parts-to-whole relationship in a dataset, but they are most effective with a small number of categories.
#------------

categories = ["Freshmen", "Sophomores", "Juniours", "Seniors"]
values = np.array([300, 250, 275, 225])
colors = ["red", "orange", "blue", "green"]

plt.pie(values, labels=categories, autopct="%1.1f%%", colors = colors,
        explode = [0,0,0,0.1], shadow = True, startangle =90) # autopct is for the percentage sign, explode is to make the a slice in the pie.
plt.title("College")
plt.show()
```

```python
# Scatter Plot: 
# A graph that displays values for two numeric variables as dots, where the position of each dot represents the values for an individual data point
# It is used to observe and visualize the relationship, or correlation, between these variables, helping to identify patterns like positive, negative, or no correlation.
# The independent variable is typically plotted on the x-axis (horizontal) and the dependent variable on the y-axis (vertical)

x = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8])
y = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87])

plt.scatter(x, y, color="red", alpha=0.8, s=100, label="Class A") # s for size, alpha for transparency

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.legend() # Shows the legend (labels)

plt.show()
```

```python
# Histograms:
# A visual representation of distributation of quantitative data.
# They group values into bins (intervals) and counts how many fall in each range.

scores = np.random.normal(loc=80, scale = 10, size = 100) # Setting the median score with loc. Scale is for how far away the scores will deviate from the center(80)
scores  = np.clip(scores, 0, 100) # Setting bounds, minimum and maximum
plt.hist(scores, bins=10, color = "lightgreen", edgecolor = "black")

plt.title("Exam Scores")

plt.xlabel('Score')
plt.ylabel("Students")

plt.show()
```

```python
# Sub plots:
# figure =  entire canvas
# ax = a single plot (subplot)
figure, axes = plt.subplots(2,2)

x = np.array([1,2,3,4,5])

axes[0, 0].plot(x, x*2, color = "red")
axes[0, 0].set_title("x*2")

axes[0, 1].bar(x, x**2, color = "blue") # can also add other charts.
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x, x**3, color = "green")
axes[1, 0].set_title("x**3")

axes[1, 1].plot(x, x**4, color="purple")
axes[1, 1].set_title("x**4")

plt.tight_layout() # To fit everything in place without overlapping.
plt.show()
```
