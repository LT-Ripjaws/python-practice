```python
import pandas as pd # Importing the pandas library.
print("Pandas-version",pd.__version__)
```

```python
# Series - A Pandas 1-Dimensional labeled array that can hold any data type.
# Like a single column in a spreadsheet (1-Dimensional)
# A series is a column and a DataFrame is a multidimensional table made up of collection of series.

data = [100, 102, 104]
print(data)
print(type(data)) # A python list. We can also use numpy to create a one dimensional array

print() # space

data.insert(0,'Numbers')

series = pd.Series(data) # Converted into a pandas series.
print(series)
print(type(series))

print(series.loc[1]) # Accessing element using label-based indexing.
print(series.iloc[1]) # Accessing element using integer-based indexing.

series.loc[3] = 300 # Accessing index 3, and inserting a new value.
print(series)

print()

data = [100, 102, 104, 200, 202, 300]
prefix = 'Item_' 
string_index  = [f'{prefix}{i+1}' for i in range(len(data))] # Adding a custom index for our series using list comprehension.
series = pd.Series(data, index=string_index)
print(series)

print(series[series>200]) #printing values greater than 200

print()

calories = {"Day 1": 1750, # Making a dictionary
            "Day 2": 2100,
            "Day 3": 1700}
series = pd.Series(calories) # Since dictionaries are key:value pairs, the labels are the keys automatically.
print(series)

new_dict = {"Day": "Calories", **calories} # Here we add a new value at the start of the dictionary.
series = pd.Series(new_dict)
print(series)
```

```python
# DataFrame = A tabular data structure with rows AND columns. (2-Dimensional)
# Similar to an Excel SpreadSheet

# Creating DataFrame Using Dictionary
data = {"Name": ["Spongebob", "Patrick", "Squidward"],
        "Age": [30, 35, 50]}

df = pd.DataFrame(data , index=['emplyee1', 'emplyee2', 'emplyee3']) # A pandas dataframe. Here the keys became the columns and the values the value in each row.
sf = pd.Series(data) # for a series, the keys become the labels.
print(df,'\n')
print(sf)

# Adding a new column
df['Job'] = ["Cook", "N/A", "Cashier"]
print(df)

# Adding a new row
new_row = pd.DataFrame([{"Name": "Sandy",
                         "Age":28,
                         "Job": "Engineer"},
                         {"Name": "Mr.Krabs",
                         "Age":50,
                         "Job": "Owner"}],
                         index = ["emplyee4", "emplyee5"]) # add dictionaries for as many rows needed.
df = pd.concat([df, new_row])
print(df)

# Creating DataFrames from List of Lists
data = [
    ['Chinmoy', 'Australia', 'Perth'], 
    ['David', 'UK', 'London'],
    ['John', 'Sweden', 'Stockholm']
]
df = pd.DataFrame(data, columns=['Names','Country','City'])
print(df)

# Creating DataFrames from a List of Dictionaries
data = [
    {'Name': 'Chinmoy', 'Country': 'Australia', 'City': 'Perth'},
    {'Name': 'David', 'Country': 'UK', 'City': 'London'},
    {'Name': 'John', 'Country': 'Sweden', 'City': 'Stockholm'}]
df = pd.DataFrame(data)
print(df)

#Check the Web_Scraping.py to see another way of inserting Columns and then rows to a DataFrame.
```

```python
# Filtration

tall_pokemon = df[df["Height"] >= 2] # syntax: df[indexing condition]. Filtering all pokemon with height greater than or equal to 2.
print(tall_pokemon)

heavy_pokemon = df[df["Weight"] >= 300] # Filtering all pokemon with weight greater than or equal to 100.
print(heavy_pokemon)

legendary_pokemon = df[df["Legendary"] == True] # Filtering all legendary pokemon.
print(legendary_pokemon)

water_pokemon = df[(df["Type1"] == "Water") | (df["Type2"] == "Water") ] # Filtering all water type pokemon.
print(water_pokemon)
```

```python
# Aggregation functions for whole DataFrame
print(df.mean(numeric_only=True)) # Mean of the columns with numeric data only.
print(df.sum(numeric_only=True)) # Sum of the columns with numeric data only.
print(df.min(numeric_only=True)) # Minimum value of the columns with numeric data only.
print(df.max(numeric_only=True)) # Maximum value of the columns with numeric data only.
print(df.count()) # Count of non-null values in each column.

# Single Column Aggregation
print(df["Height"].mean()) 
print(df["Height"].sum()) 
print(df["Height"].min()) 
print(df["Height"].max()) 
print(df["Height"].count())

# GroupBy function:
group = df.groupby('Type1') # Grouping the dataframe by Type1 column. 
print(group["Height"].mean()) # Getting the mean height of each group.
```

```python
# Data Cleaning - Handling Missing Data
# 1. Drop irrelevant data
df_new = df.drop(columns=['Legendary', "No"]) # Dropping the 'Legendary' column as it is not relevant for our analysis.
print(df_new)

# 2. Handling missing data
# df_new = df.dropna(subset=['Type2']) # Dropping rows where 'Type2' column has missing data.
# print(df_new)
df_new = df.fillna({'Type2': 'Unknown'}) # Filling missing data in 'Type2' column with 'Unknown'.
print(df_new)

# 3. Fix inconsistent values
df_new["Type1"] = df_new["Type1"].replace({"Grass": "GRASS",
                                           "Fire": "FIRE"})
print(df_new)

# 4. Standardize Text
df_new["Name"] = df_new["Name"].str.upper()
print(df_new)

# 5. Fix Data Types
df_new["Legendary"] = df["Legendary"].astype(bool)
print(df_new)

# 6. Remove Duplicate Values
df_new = df.drop_duplicates()
```
